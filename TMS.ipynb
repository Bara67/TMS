```python
zip_file_path = '/content/drive/MyDrive/dataset/images.zip'
extract_folder = '/content/drive/MyDrive/extracted_dataset/'

# Membuat folder tujuan jika belum ada
if not os.path.exists(extract_folder):
    os.makedirs(extract_folder)

# Mengekstrak file ZIP
if os.path.exists(zip_file_path):
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(extract_folder)

print("File ZIP berhasil diekstrak.")

# Fungsi untuk memproses gambar
def preprocess_image(image_path, img_size=(128, 32)):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"Gagal memuat gambar: {image_path}")
        return None
    img = cv2.resize(img, img_size)
    img = img / 255.0
    img = img.reshape((*img_size, 1))
    return img

# Fungsi untuk memuat data
def load_data(folder_path):
    images = []
    for filename in os.listdir(folder_path):
        if filename.endswith(('.jpg', '.png', '.jpeg')):
            image_path = os.path.join(folder_path, filename)
            img = preprocess_image(image_path)
            if img is not None:
                images.append(img)
    if len(images) == 0:
        print(f"Tidak ada data gambar valid di {folder_path}.")
    return np.array(images)

# Memuat data pelatihan dan validasi
train_images = load_data(extract_folder)
val_images = load_data(extract_folder)

if train_images.size == 0 or val_images.size == 0:
    raise ValueError("Data gambar kosong. Periksa path dan format gambar.")

print(f'Train images shape: {train_images.shape}')
print(f'Validation images shape: {val_images.shape}')

# Dummy labels untuk training
num_classes = 80
dummy_labels = np.zeros((train_images.shape[0], num_classes))
val_labels = np.zeros((val_images.shape[0], num_classes))

# Membuat model CRNN
def build_crnn_model(input_shape, num_classes):
    inputs = Input(shape=input_shape)
    x = Conv2D(32, (3, 3), activation='relu', padding='same')(inputs)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)
    x = Reshape((-1, x.shape[-1]))(x)
    x = Bidirectional(LSTM(128, return_sequences=True))(x)
    x = Bidirectional(LSTM(128, return_sequences=True))(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = tf.keras.models.Model(inputs, outputs)
    return model

# Fungsi CTC loss
def ctc_loss(y_true, y_pred):
    input_length = tf.reduce_sum(tf.ones_like(y_pred[:, :, 0]), axis=1)
    label_length = tf.reduce_sum(tf.ones_like(y_true), axis=1)
    return tf.reduce_mean(tf.keras.backend.ctc_batch_cost(y_true, y_pred, input_length, label_length))

# Menyiapkan model
model = build_crnn_model(input_shape=(32, 128, 1), num_classes=num_classes)
model.compile(optimizer='adam', loss=ctc_loss)

# Melatih model
model.fit(train_images, dummy_labels, validation_data=(val_images, val_labels), epochs=50, batch_size=32)
```
